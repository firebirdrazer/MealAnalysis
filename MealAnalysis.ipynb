```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_excel('MealAnalysis.xlsx','data')
data.head(10)
```

We can first load data and check it if there's anything happens unexcepted.
Just like the 8th row has null data, then I'll drop it for cleansing.

```python
data1 = data.drop(data[data['Type'].isnull()].index)
data1.head(10)
```

Then we can simplify the data from gender for 2 subsets, data_male and data_female.

```python
data_male = data1[data1['gender']=='male']
data_female = data1[data1['gender']=='female']
```

Here I ignore the differences between meal types for simplifying the process, thus we could make more detailed analysis by drilling down the types.
The data range is obviously wider than imagine, so at the first we could give them a normalized process.
We could make data precess on 1 subset first, here I choose the data_male.

From the correlation matrix, we can find that the EER and targets of protein, fat and carbo are highly correlated, thus we only keep the EER here. Besides, the calories E is also highly correlated to Protein, Fat and Carbo. But here I'll pop the E column because the calories are calculated from those nutrients.

```python
df = data_male[data_male.columns[2:-2]]
ndf = (df-df.mean())/df.std()
corr = ndf.corr()
corr
```

```python
ndf.drop(["P target(15%)[g]","F target(25%)[g]","C target(60%)[g]","E[kcal]"],axis = 1, inplace = True)
Score = data_male[data_male.columns[-1]]
```

As I remove the correlated features and there are only normalized features, I applied a simple linear regression model here to find out the relationship between the features and scores.

We can find the weights of each feature here, which means the age and height are not necessary features here. The Fat is the only negative feature here, which means the more fat, the less score.

```python
linearModel=LinearRegression()
linearModel.fit(ndf,Score)
print(ndf.columns)
print(linearModel.coef_)
print(linearModel.intercept_)
```

```python
mse = np.mean((linearModel.predict(ndf) - Score) ** 2)
r_squared = linearModel.score(ndf, Score)
adj_r_squared = r_squared - (1 - r_squared) * (ndf.shape[1] / (ndf.shape[0] - ndf.shape[1] - 1))
print(mse, r_squared, adj_r_squared)
```

I also evaluate the linear regression model, for MSE, R^2 and adjusted R^2 are all lower than 0.5, this model could be modified more. 

```python
df = data_female[data_female.columns[2:-2]]
ndf = (df-df.mean())/df.std()
ndf.drop(["P target(15%)[g]","F target(25%)[g]","C target(60%)[g]","E[kcal]"],axis = 1, inplace = True)
Score = data_female[data_female.columns[-1]]
linearModel=LinearRegression()
linearModel.fit(ndf,Score)
print(ndf.columns)
print(linearModel.coef_)
print(linearModel.intercept_)
mse = np.mean((linearModel.predict(ndf) - Score) ** 2)
r_squared = linearModel.score(ndf, Score)
adj_r_squared = r_squared - (1 - r_squared) * (ndf.shape[1] / (ndf.shape[0] - ndf.shape[1] - 1))
print(mse, r_squared, adj_r_squared)
```

From the same analysis applied to female subset, the result is almost the same.
